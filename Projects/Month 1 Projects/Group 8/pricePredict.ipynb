```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("/content/drive/MyDrive/datasets/californiahouseprice.csv")
df.head(5)
```

```python
df.shape
```

```python
df["ocean_proximity"].unique()
```

```python
df1 = df.drop(["longitude","latitude","population"], axis = 1)
df1.head()
```

```python
df1.shape
```

```python
df1.isnull().sum()
```

```python
df2 = df1.dropna()
df2.isnull().sum()
```

```python
df2.shape
```

```python
df2.groupby("ocean_proximity")["ocean_proximity"].agg("count")
```

```python
df2.loc[df2["ocean_proximity"]=="ISLAND"]
```

```python
df3 = df2[~(df2["ocean_proximity"]=="ISLAND")]
df3.shape
```

```python
df3["ocean_proximity"].unique()
```

```python
dummies =  pd.get_dummies(df3["ocean_proximity"]).astype("int")
dummies
```

```python
df4 = df3.drop(["ocean_proximity"],axis = 1)
df4.head()
```

```python
df5 = pd.concat([df4,dummies],axis = 1)
df5.head()
```

```python
df5.rename(columns = {"<1H OCEAN":"<1HR FROM OCEAN"},inplace = True)
```

```python
df5.head()
```

```python
df5.columns = df5.columns.str.replace('_', ' ')
df5.head()
```

```python
X = df5.drop("median house value",axis = 1)
Y = df5["median house value"]
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
scaler1 = scaler.fit(X)
```

```python
X_scaled = scaler.transform(X)
X_scaled
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X_scaled, Y, test_size=0.2, random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
model.fit(X_train,Y_train)
```

```python
Y_pred = model.predict(X_test)
Y_pred[:10]
```
